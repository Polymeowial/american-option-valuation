```python

import numpy as np
import matplotlib.pyplot as plt

S0 = 110; L=95; K=105
r = 0.05; sig=0.2
T = 1.0
NStep = 252
dt = T / NStep
t = np.linspace(0, T, NStep)

def v(S_t):
    if S_t <= L:
        return K - S_t
    else:
        return (K - L) * (S_t/L)**(-2*r/(sig**2))


dW = np.sqrt(dt) * np.random.standard_normal(size=NStep) 
W = np.cumsum(dW)
S = S0 * np.exp((r - 0.5 * sig**2) * t + sig * W)













val = [v(s) for s in S]

val_stopped = np.array([np.nan] * len(S))
for i in range(len(S)):
    val_stopped[i] = v(S[i])

    if S[i] <= L:
        val_stopped[i] = v(S[i])
        print(f"First time stock price reach {L} at time {i}")
        break

for j in range(i+1, len(S)):
    val_stopped[j] = val_stopped[i]


fig, (ax, ax1, ax2) = plt.subplots(1,3, figsize=(15,5))
ax.plot(t, S)
ax.axvline(x=t[i], color='red', linestyle='--', linewidth=2)
ax.set_title("Stock price")

ax1.plot(t, val)
ax1.axvline(x=t[i], color='red', linestyle='--', linewidth=2)
ax1.set_title("Amer put price")

ax2.plot(t, val_stopped)
ax2.axvline(x=t[i], color='red', linestyle='--', linewidth=2)
ax2.set_title("Amer put price (stopped)")

plt.show(); plt.close()
```
